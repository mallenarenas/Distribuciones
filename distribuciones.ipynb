<h2> Librerias necesarias </h2>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t, chi2, bernoulli
```

<h2> Distribuciones continuas </h2>

<h3> 1. La Distribución Normal </h3>

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$
$\mu \in (-\infty, \infty)$ y $\sigma > 0$
<h4> 1.1 Interpretación de la media en la gráfica </h4>

```python
# Crear una sucesión de números entre -9 y 9, aumentando en 0.05.
x = np.arange(-5, 5, 0.05)

# Variando la Desviación Estándar.
y1 = norm.pdf(x, loc=0, scale=1)
y2 = norm.pdf(x, loc=2, scale=1)
y3 = norm.pdf(x, loc=-2, scale=1)

# Gráfica de la densidad normal
plt.plot(x, y1, color="red", linewidth=2, label="mu = 0")
plt.plot(x, y2, color="black", label="mu = 2")
plt.plot(x, y3, color="blue", label="mu = -2")

# Agregar la leyenda
plt.legend(loc="upper right")

# Etiquetas de los ejes y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Distribución Normal variando la media")

# Mostrar el gráfico
plt.show()
```

<h4> 1.2 Interpretación de la desviación estándar en la gráfica </h4>


```python
# Crear una sucesión de números entre -9 y 9, aumentando en 0.05.
x = np.arange(-9, 9, 0.05)

# Variando la Desviación Estándar.
y1 = norm.pdf(x, loc=0, scale=0.7)
y2 = norm.pdf(x, loc=0, scale=1)
y3 = norm.pdf(x, loc=0, scale=3)

# Gráfica de la densidad normal
plt.plot(x, y1, color="red", linewidth=2, label="sd = 0.7")
plt.plot(x, y2, color="black", label="sd = 1")
plt.plot(x, y3, color="blue", label="sd = 3")

# Agregar la leyenda (viñeta)
plt.legend(loc="upper right")

# Etiquetas de los ejes y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Distribución Normal variando sd")

# Mostrar el gráfico
plt.show()
```

<h3> 2. La Distribución T-Student </h3>

$$f(x| \nu) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu\pi} \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{x^2}{\nu}\right)^{-\frac{\nu + 1}{2}}$$
donde $\Gamma(x)$ es la función gamma y
$$\Gamma(x) = \int_{0}^{\infty} t^{x-1} e^{-t} dt $$

```python
# Crear una sucesión de números entre -5 y 5, aumentando en 0.05.
x = np.arange(-5, 5, 0.05)

# Variando la Desviación Estándar.
y1 = norm.pdf(x, loc=0, scale=1)
df1=5
df2=1
y2 = t.pdf(x, df=df1)
y3 = t.pdf(x, df=df2)

# Gráfica de la densidad normal
plt.plot(x, y1, color="red", linewidth=2, label="Normal(0,1)")
plt.plot(x, y2, color="black", label="df = {}".format(df1))
plt.plot(x, y3, color="blue", label="df = {}".format(df2))

# Agregar la leyenda (viñeta)
plt.legend(loc="upper right")

# Etiquetas de los ejes y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Distribución T-Student")

# Mostrar el gráfico
plt.show()
```

<h3> 3. La Distribución Chi-cuadrado </h3>

$$f(x| k) = \frac{1}{2^{\frac{k}{2}} \Gamma\left(\frac{k}{2}\right)} x^{\frac{k}{2}-1} e^{-\frac{x}{2}}$$

donde $\Gamma(x)$ es la función gamma y
$$\Gamma(x) = \int_{0}^{\infty} t^{x-1} e^{-t} dt $$


```python
# Crear una sucesión de números entre 0.05 y 20, aumentando en 0.05.
x = np.arange(0.05, 20, 0.05)

# Variando los grados de libertad.
df1=3
df2=7
df3=10
y1 = chi2.pdf(x, df=df1)
y2 = chi2.pdf(x, df=df2)
y3 = chi2.pdf(x, df=df3)

# Gráfica de la chi-cuadrado
plt.plot(x, y1, color="red", linewidth=2, label="df= {}".format(df1))
plt.plot(x, y2, color="black", label="df = {}".format(df2))
plt.plot(x, y3, color="blue", label="df = {}".format(df3))

# Agregar la leyenda (viñeta)
plt.legend(loc="upper right")

# Etiquetas de los ejes y título
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Distribución Chi-Cuadrado")

# Mostrar el gráfico
plt.show()
```

<h3> 4. La Distribución F </h3>

$$f(x| dfn, dfd) = \frac{\sqrt{\frac{(dfn \cdot x)^{dfn} \cdot dfd^{dfd}}{(dfn \cdot x + dfd)^{dfn + dfd}}}}{x \cdot B\left(\frac{dfn}{2}, \frac{dfd}{2}\right)}$$
donde $B(x, y)$ es la función beta
$$B(x, y) = \frac{\Gamma(x) \cdot \Gamma(y)}{\Gamma(x + y)}$$

```python
from scipy.stats import f

# Grados de libertad del numerador y denominador
dfn1 = 5
dfn2 = 10
dfn3 = 20
dfd1 = 10
dfd2 = 5
dfd3 = 20

# Valores posibles de la distribución F
x = np.linspace(0.01, 5, 1000)

# Función de densidad de probabilidad (PDF) de la distribución F
y1_1 = f.pdf(x, dfn1, dfd1)
y2_1 = f.pdf(x, dfn2, dfd1)
y3_1 = f.pdf(x, dfn3, dfd1)
y1_2 = f.pdf(x, dfn1, dfd2)
y1_3 = f.pdf(x, dfn2, dfd3)


# Gráfica de la distribución F
plt.plot(x, y1_1, color="blue", linewidth=2, label="(dfn={}, dfd={})".format(dfn1,dfd1))
plt.plot(x, y2_1, color="red", linewidth=2, label="(dfn={}, dfd={})".format(dfn2,dfd1))
plt.plot(x, y3_1, color="green", linewidth=2, label="(dfn={}, dfd={})".format(dfn3,dfd1))
plt.plot(x, y1_2, color="purple", linewidth=2, label="(dfn={}, dfd={})".format(dfn1,dfd2))
plt.plot(x, y1_3, color="grey", linewidth=2, label="(dfn={}, dfd={})".format(dfn1,dfd3))

# Agregar la leyenda (viñeta)
plt.legend(loc="upper right")

plt.xlabel("Valor de F")
plt.ylabel("Probabilidad")
plt.title("Distribución F (dfn={}, dfd={})".format(dfn1, dfd1))
plt.grid(True)
plt.show()
```

```python
from scipy.stats import f

# Valores posibles de la distribución F
x = np.linspace(0.01, 5, 1000)

# Crear la figura con dos subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Gráfica de la distribución F con variación en dfn
dfn_values = [5, 10, 20]
dfd_fixed = 10
for dfn in dfn_values:
    y = f.pdf(x, dfn, dfd_fixed)
    axes[0].plot(x, y, label="(dfn={}, dfd={})".format(dfn, dfd_fixed))

axes[0].set_xlabel("Valor de F")
axes[0].set_ylabel("Probabilidad")
axes[0].set_title("Distribución F (Variación en dfn)")
axes[0].grid(True)
axes[0].legend()

# Gráfica de la distribución F con variación en dfd
dfd_values = [10, 5, 20]
dfn_fixed = 5
for dfd in dfd_values:
    y = f.pdf(x, dfn_fixed, dfd)
    axes[1].plot(x, y, label="(dfn={}, dfd={})".format(dfn_fixed, dfd))

axes[1].set_xlabel("Valor de F")
axes[1].set_ylabel("Probabilidad")
axes[1].set_title("Distribución F (Variación en dfd)")
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

<h3> 5. La distribución bernoulli </h3>

$$f(x|p) = \begin{cases}
p & \text{si } x = 1 \\
1-p & \text{si } x = 0
\end{cases}$$

```python
# Probabilidad de éxito (p) en el experimento Bernoulli
p = 0.3

# Valores posibles del experimento (éxito o fracaso)
x = [0, 1]

# Probabilidades asociadas a los valores (fracaso o éxito)
probabilidades = [1-p, p]

# Generar la distribución Bernoulli
distribucion = bernoulli(p)

# Gráfica de barras para la distribución Bernoulli
plt.bar(x, probabilidades, color="green")
plt.xticks(x, ['Fracaso (0)', 'Éxito (1)'])
plt.xlabel('Resultado del Experimento')
plt.ylabel('Probabilidad')
plt.title('Distribución Bernoulli (p={})'.format(p))
plt.show()
```

<h3>6. La distribución Binomial</h3>

$$f(k|n,p) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$$
donde $\binom{n}{k}$ es el coeficiente binomial
$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$

```python
from scipy.stats import binom

# Número de ensayos
n = 10

# Probabilidad de éxito en cada ensayo
p = 0.5

# Valores posibles de la distribución binomial (número de éxitos)
x = np.arange(0, n+1)

# Probabilidades asociadas a cada número de éxitos
probabilidades = binom.pmf(x, n, p)

# Gráfica de barras para la distribución binomial
plt.bar(x, probabilidades, color="purple")
plt.xticks(x)
plt.xlabel('Número de Éxitos')
plt.ylabel('Probabilidad')
plt.title('Distribución Binomial (n={}, p={})'.format(n, p))
plt.show()
```
